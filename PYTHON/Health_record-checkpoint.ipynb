# data clean and preprocessing of health_record

```python
import pandas as pd
# Reading the CSV file into a DataFrame
df = pd.read_csv('HealthRecords.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

# Exploratory Data Analysis (EDA) stage

```python
#Basic Statistics: Understand the basic statistics of each column.
summary = df.describe(include='all')
print(summary)
```

```python
# Get the value counts for MedicalHistory
MedicalHistory_counts = df['MedicalHistory'].value_counts()
print("MedicalHistory Distribution:")
print(MedicalHistory_counts)
print("\n")  # Just to add a new line for separation
```

```python
# Get the value counts for CurrentMedications
CurrentMedications_counts = df['CurrentMedications'].value_counts()
print("CurrentMedications Distribution:")
print(CurrentMedications_counts)
print("\n")  # Just to add a new line for separation
```

```python
# Get the value counts for DiseaseTypes
DiseaseTypes_counts = df['DiseaseTypes'].value_counts()
print("DiseaseTypes Distribution:")
print(DiseaseTypes_counts)
print("\n")  # Just to add a new line for separation
```

```python
#to understand the distribution.
import matplotlib.pyplot as plt

# Bar plot for MedicalHistory
MedicalHistory_counts = df['MedicalHistory'].value_counts()
MedicalHistory_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Workers have different MedicalHistory') #none is mean they don't have Medical issue before, keep it
plt.xlabel('MedicalHistory')
plt.ylabel('Number of Workers')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Bar plot for CurrentMedications
CurrentMedications_counts = df['CurrentMedications'].value_counts()
CurrentMedications_counts.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('CurrentMedications for workers')
plt.xlabel('CurrentMedications')
plt.ylabel('Number of Workers have same Medications')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Bar plot for DiseaseTypes
DiseaseTypes_counts = df['DiseaseTypes'].value_counts()
DiseaseTypes_counts.plot(kind='bar', color='yellow', edgecolor='black')
plt.title('DiseaseTypes for workers')
plt.xlabel('DiseaseTypes')
plt.ylabel('Number of Workers have same DiseaseTypes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

# Pie chart

```python
print(df.info())
```

```python
# count None and others
value_counts = df['MedicalHistory'].value_counts()
none_count = value_counts.get('None', 0)  # get number of None(who are not have medical issue before), if not set as 0
other_count = df.shape[0] - none_count  # the number of people have medical issue before

# prepare the things to draw the pie chart
labels = ['None', 'Other']
sizes = [none_count, other_count]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # highlight the None

# draw the pie
#use the autopct to show the percentage in the chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140) #shadow make the graph looks good 
plt.axis('equal')  # keep pie is circle
plt.title('Distribution of Medical History (None vs. Other)')
plt.show()
```

```python
# relationship between MedicalHistory and CurrentMedications
cross_tab = pd.crosstab(df['MedicalHistory'], df['CurrentMedications'])

# stack barchart
plt.figure(figsize=(12, 6))
cross_tab.plot(kind='bar', stacked=True, colormap='Set3', ax=plt.gca())
plt.title('Medical History vs Current Medications')
plt.xlabel('Medical History')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Current Medications', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
