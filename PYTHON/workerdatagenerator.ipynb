```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Number of records
n = 1000

# Occupations and their ratios
occupations = ['Miner', 'Engineer', 'Geologist', 'Surveyor', 'Mechanic']
occupation_ratios = [0.4, 0.25, 0.15, 0.1, 0.1]

# Reasons for leaving and their ratios
reasons_for_leaving = ['Safety Concerns', 'Better Opportunities Elsewhere', 'Family Reasons', 'Health Issues', 'Relocation', 'Dissatisfaction with Pay', 'End of Contract/Project', 'Retirement', 'Company Downsizing']
reason_ratios = [0.25, 0.2, 0.15, 0.15, 0.1, 0.05, 0.05, 0.03, 0.02]

# Initialize empty DataFrame
df = pd.DataFrame(columns=['WorkerID', 'Age', 'Occupation', 'IsCurrentEmployee', 'DateJoined', 'DateLeft', 'ReasonForLeaving'])

for i in range(n):
    # Assigning WorkerID
    worker_id = i + 1

    # Randomly assign Age between 25 to 60
    age = random.randint(25, 60)

    # Randomly pick Occupation based on the given ratios
    occupation = random.choices(occupations, weights=occupation_ratios, k=1)[0]

    # Randomly determine if the person is a current employee
    is_current_employee = random.choice([True, False])

    # Randomly pick DateJoined within the last 10 years
    date_joined = datetime.now() - timedelta(days=random.randint(0, 3650))

    # Initialize DateLeft and ReasonForLeaving
    date_left = None
    reason_for_leaving = None

    # If the person is not a current employee, fill in DateLeft and ReasonForLeaving
    if not is_current_employee:
        # Randomly pick DateLeft after DateJoined
        date_left = date_joined + timedelta(days=random.randint(0, (datetime.now() - date_joined).days))

        # Randomly pick ReasonForLeaving based on the given ratios
        reason_for_leaving = random.choices(reasons_for_leaving, weights=reason_ratios, k=1)[0]

    # Append the generated record to the DataFrame
    df = df.append({'WorkerID': worker_id, 'Age': age, 'Occupation': occupation, 'IsCurrentEmployee': is_current_employee, 'DateJoined': date_joined, 'DateLeft': date_left, 'ReasonForLeaving': reason_for_leaving}, ignore_index=True)

# Export the DataFrame to a CSV file
df.to_csv('worker_dataset.csv', index=False)
```
