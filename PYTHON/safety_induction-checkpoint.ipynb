# data clean and preprocessing of safety_induction

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Reading the CSV file into a DataFrame
df = pd.read_csv('safety_induction_dataset.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
print(df.info())
```

# Exploratory Data Analysis (EDA) stage

```python
#Basic Statistics: Understand the basic statistics of each column.
summary = df.describe(include='all')
print(summary)
```

```python
#check all the date is correct format
from datetime import datetime
invalid_dates = []

for date_str in df['InductionDate']:
    try:
        datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S.%f')
    except ValueError:
        invalid_dates.append(date_str)

print("Invalid dates:")
for date_str in invalid_dates:
    print(date_str)
```

```python
# check is there any null value
df.isnull().sum()  
```

```python
# Get the value counts for InductionType
InductionType_counts = df['InductionType'].value_counts()
print("InductionType Distribution:")
print(InductionType_counts)
print("\n")  # Just to add a new line for separation
```

```python
# Bar plot for InductionType
InductionType_counts = df['InductionType'].value_counts()
InductionType_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Workers have different InductionType') #none is mean they don't have Medical issue before, keep it
plt.xlabel('InductionType')
plt.ylabel('Number of Workers')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# histgram
# number of different score we have
# like we can see that there is 100 more people get score 0-5
plt.hist(df['InductionScore'], bins=20)
plt.xlabel('Induction Score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# the total average score
mean_score = df['InductionScore'].mean()

# show the average score for different induction type
group_stats = df.groupby('InductionType')['InductionScore'].mean()
```

```python
# check is pass or not
pass_count = (df['InductionScore'] >= 50).sum() 
fail_count = (df['InductionScore'] < 50).sum()  

# result
print(f"pass: {pass_count}")
print(f"no pass: {fail_count}")
```

```python
# check is pass or not
pass_count = (df['InductionScore'] >= 50).sum() 
fail_count = (df['InductionScore'] < 50).sum()  

# data we need 
labels = ['Pass', 'Fail']
sizes = [pass_count, fail_count]
colors = ['green', 'red']
explode = (0.1, 0)  

# pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Pass vs. Fail Inductions')
plt.show()
```

```python
# Create histograms and kernel density estimates
sns.distplot(df['InductionScore'], kde=True)
plt.xlabel('Induction Score')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# box plot
# show the more detail about score for each type
sns.boxplot(data=df, x='InductionType', y='InductionScore')
plt.xlabel('Induction Type')
plt.ylabel('Induction Score')
plt.xticks(rotation=45)
plt.show()
```
